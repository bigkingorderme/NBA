```python
# 尝试获取所有球队的信息

def getLineUp(team):
    get_lineup_url = 'https://data.zhibo8.cc/manage/public/app.php?_url=/nba_v2/team&random=&teamName='
    return requests.get(get_lineup_url + team).json()

getLineUp('火箭')
```

```python
# 将球员信息存到一个dataframe中

player = DataFrame()

for i in range(len(team)):
    # 加上球队名
    temp = DataFrame(team.iloc[i]['lineup']['data']['player']['info']['list'])
    temp['球队名称'] = team.iloc[i]['lineup']['data']['team']['teamName']
    temp['球队logo'] = team.iloc[i]['lineup']['data']['team']['logo']
    player = player.append(temp)
```

```python
player_2 = DataFrame()

for i in range(len(team)):
    player_2 = player_2.append(team.iloc[i]['lineup']['data']['player']['data']['list'])
```

```python
player_2.shape
```

```python
player_2.head()
```

```python
# 合并两个球员表

player['playerId'] = player['playerId'].astype(int)  # 两个key必须是相同类型

player_all = pd.merge(player, player_2, on='playerId',how='left')
player_all.info()
```

```python
# 很多数字的字段是object类型，需要转化成int,具体分析到的时候再做转换吧，比如先分析工资
# 获取球员当赛季薪资

import re

def getSalaryInt(salary):
    N = ''
    number = re.findall(r'\d+', salary)
    for i in range(len(number)):
        N = N+number[i]
    if N == '':
        return 0
    else:
        return int(N)
    
getSalaryInt(player_all.iloc[1]['当赛季薪资'])
```

```python
player_all['当赛季薪资'] = player_all['当赛季薪资'].apply(getSalaryInt)
player_all['当赛季薪资']
```

```python
# 部分球员薪资为0，统计的还是应该剔除

player_all[player_all['当赛季薪资'] != 0].describe()

#平均 774.64万美元
```

```python
player_salary = player_all[player_all['当赛季薪资'] != 0]
player_salary.groupby(by='球队名称').mean()['当赛季薪资'].sort_values(ascending=False)
```
